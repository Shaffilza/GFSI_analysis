```python
# read excel Covid Cases
import pandas as pd
df = pd.read_excel('Covid19_Case.xlsx')
df.head() 
```

```python
# select only last date by year, group by location

df2=df.groupby([pd.DatetimeIndex(df.date).to_period('Y'),'location']).nth([-1])
df2
```

```python
# dataframe to excel
df2.to_excel("output.xlsx")
```

```python
# read excel output for updated covid cases data
import pandas as pd
df4 = pd.read_excel('output.xlsx')
df4.head() 
```

```python
# read excel GFSI_2020
import pandas as pd
df3 = pd.read_excel('GFSI_2020.xlsx')
df3.head() 
```

```python
# read excel GFSI_2021
import pandas as pd
df5 = pd.read_excel('GFSI_2021.xlsx')
df5.head() 
```

```python
# read excel GFSI_2019
import pandas as pd
df6 = pd.read_excel('GFSI_2019.xlsx')
df6.head() 
```

```python
#merge GFSI_2020 and GFSI_2021 
merge1=df3.merge(df5, how='outer')
merge1
```

```python
#merge GFSI_2020 and GFSI_2021 and GFSI_2019 
merge2=df6.merge(merge1, how='outer')
merge2
```

```python
# MERGE COVID DATASET AND GFSI


merge3=merge2.merge(df4, how='inner', on=['YEAR', 'COUNTRY'])
merge3
```

```python
# fillna (remove NAN and insert 0)
removeNAN=merge3.fillna(value=0)
```

```python
# dataframe to excel to check
removeNAN.to_excel("output2.xlsx")
```
