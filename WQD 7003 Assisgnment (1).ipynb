## Overview and Motivation
Amidst the two years of the COVID-19 pandemic, the global food systems have greatly disrupted, and this food insecurity crisis continue to escalate and worsen as it struggles to feed its 7.9 billion people. 
#### UN estimated that between 720 million and 811 million of people around the world suffer from some degree of hunger in 2020, which also depicted that about 70 million to 161 million more people suffer from hunger in 2020 when compared to those in 2019. (Whiting, 2022) 
#### With the explosion of Ukraine-Russia war, it seems that this is the last straw that breaks the camel’s back as Ukraine and Russia has both accounted for a total of 12% of the food calories traded globally. 
This undeniably has put the already worsen conditions into a deeper concern for the international community. 
#### According to World Food Summit (1996), food security is defined as a condition in which all people are able to physically and economically given access to enough, safe and healthy food to satisfy their dietary needs and food preferences for a healthy life. 
#### In 2012, the Global Food Security Index (GFSI) was first introduced by The Economist and updated every year. To formulate the model, 58 unique indicators are included to measure the degree of food security across different countries in which four main aspects are taken into consideration which are food affordability, availability, quality and safety and lastly natural resources and resilience. 
Therefore, after referring to the GFSI, our group has decided to use the datasets obtained from the official website of GFSI to explore the average 10 years of GFSI for 113 countries and identify the countries with minimum and maximum parameters of GFSI.

```python
def linearRegressionScore(x,y):
    LR = LinearRegression() 
    LR.fit(x,y)
    return LR.score(x,y) 

def correlationNetwork(df, minCorr=-1.1, edgeLabels=True, figsize=20):
    corr = df.corr()
 
    # Transform it in a links data frame (3 columns only):
    corr = corr.stack().reset_index()
    corr.columns = ['var1', 'var2', 'value']

    # Keep only correlation over a threshold and remove self correlation (cor(A,A)=1)
    corr=corr.loc[(corr['value'] > minCorr) & (corr['var1'] != corr['var2'])] 
    corr['value'] = round(corr['value'],2)

    # Build your graph
    G=nx.from_pandas_edgelist(corr, 'var1', 'var2',True) 

    # build network layout
    edges,weights = zip(*nx.get_edge_attributes(G,'value').items())
    weights = tuple([(1+abs(x))**2 for x in weights])
    pos = nx.kamada_kawai_layout(G,weight='value')

    # Plot the network:
    plt.figure(1,figsize=(figsize,figsize))  
    nx.draw(G, pos, with_labels=True, node_color='orange',width=weights, edge_color = weights, edge_cmap = plt.cm.GnBu)
    if (edgeLabels):
        nx.draw_networkx_edge_labels(G, pos, edge_labels=nx.get_edge_attributes(G,'value'), font_color='red')
    plt.show()
```

### Import Necessary Libraries

```python
def mannwhitneyu_tests(x,y,cols,sig=0.005,ls=True):
    res = [(col,) + stats.mannwhitneyu(x[col],y[col]) for col in cols]
    df1 = pd.DataFrame(res, columns=['column', 't', 'p']).sort_values(by=['t'],ascending = not ls)
    if ls:
        return df1[df1.p < sig]
    else:
        return df1[df1.p > sig]

def kruskal_tests(*bins,cols,sig=0.005,ls=True):
    res = [(col,) + stats.kruskal(*[b[col] for b in bins]) for col in cols]
    df1 = pd.DataFrame(res, columns=['column', 't', 'p']).sort_values(by=['t'],ascending = not ls)
    if ls:
        return df1[df1.p < sig]
    else:
        return df1[df1.p > sig]
  
def anova_tests(*bins,cols,sig=0.005,ls=True):
    res = [(col,) + stats.f_oneway(*[b[col] for b in bins]) for col in cols]
    df1 = pd.DataFrame(res, columns=['column', 't', 'p']).sort_values(by=['t'],ascending = not ls)
    if ls:
        return df1[df1.p < sig]
    else:
        return df1[df1.p > sig]
    
def strip_box(data, anova_df, x,ncols=5):        
    n = len(anova_df.index)
    f, axs = plt.subplots(figsize=(10, 10 if n < 10 else n), ncols=ncols, nrows=math.ceil(n/ncols))
    plt.subplots_adjust(top=ncols/5,wspace=(ncols*3)/20)
    for i in range(n):
        sns.boxplot(x=x, y=anova_df['column'].iloc[i],data=data, ax=axs[math.floor(i/ncols), i%ncols])
        sns.stripplot(x=x, y=anova_df['column'].iloc[i],data=data, ax=axs[math.floor(i/ncols), i%ncols],color=".3")
        
def hists(data, anova_df, x,ncols=3,kde_only=False):        
    n = len(anova_df.index)
    f, axs = plt.subplots(figsize=(10, 10 if n < 10 else n), ncols=ncols, nrows=math.ceil(n/ncols))
    plt.subplots_adjust(top=ncols/3, wspace=ncols/3, hspace=ncols/12)
    for i in range(n): 
        if(kde_only):
            sns.kdeplot(data=data,x=df1['column'].iloc[i],hue=x, ax=axs[math.floor(i/ncols), i%ncols]) 
        else:
            sns.histplot(data=data,x=df1['column'].iloc[i],hue=x,kde=True, ax=axs[math.floor(i/ncols), i%ncols]) 
```

### 1st Test :  Wilcoxon signed-rank test 
The Wilcoxon signed-rank test tests the null hypothesis that two related paired samples come from the same distribution. In particular, it tests whether the distribution of the differences x - y is symmetric about zero. It is a non-parametric version of the paired T-test.

```python
#define country with each factors
# subset dataframe according to country and main indicator, 
# Affordability , Availability , Quality & Safety ,Natural Resource & Resilience

congo_aff=new_GFSI_df.iloc[24,3:17]
congo_avv=new_GFSI_df.iloc[24,17:39]
congo_qs=new_GFSI_df.iloc[24,39:55]
congo_nr=new_GFSI_df.iloc[24,55:82]
ireland_aff=new_GFSI_df.iloc[498,3:17]
ireland_avv=new_GFSI_df.iloc[498,17:39]
ireland_qs=new_GFSI_df.iloc[498,39:55]
ireland_nr=new_GFSI_df.iloc[498,55:82]
```

```python
# Wilcoxon-Signed Rank Test between Congo and Ireland (AFFORDABILITY)
# Create data
import scipy.stats as stats
  
# conduct the Wilcoxon-Signed Rank Test
w_aff,p_aff=stats.wilcoxon(congo_aff, ireland_aff, alternative='less', zero_method="zsplit")
print("w= ",w_aff,"   p-value= ",p_aff)

if p_aff < 0.05:
  print(
      'p-value is less than 0.05, reject null hypothesis.'
      'We have sufficient evidence to say that the mean is not equal between both countries'
      
  )
else:
    print(
       'p-value is larger than 0.05, do not reject null hypothesis.'
       'There is a no significant different between both countries.'
     
  )
```

```python
# Wilcoxon-Signed Rank Test between Congo and Ireland (AVAILABILITY)
# Create data
import scipy.stats as stats
  
# conduct the Wilcoxon-Signed Rank Test
w_avv,p_avv=stats.wilcoxon(congo_avv, ireland_avv,alternative='less', zero_method="zsplit")
print("w= ",w_avv,"   p-value= ",p_avv)

if p_avv < 0.05:
  print(
      'p-value is less than 0.05, reject null hypothesis.'
      'We have sufficient evidence to say that the mean is not equal between both countries'
      
  )
else:
    print(
       'p-value is larger than 0.05, do not reject null hypothesis.'
       'There is a no significant different between both countries.'
     
  )

```

```python
# Wilcoxon-Signed Rank Test between Congo and Ireland (QUALITY AND SAFETY)
# Create data
import scipy.stats as stats
  
# conduct the Wilcoxon-Signed Rank Test
w_qs,p_qs=stats.wilcoxon(congo_qs, ireland_qs, alternative='less',  zero_method="zsplit")
print("w= ",w_qs,"   p-value= ",p_qs)

if p_qs < 0.05:
  print(
      'p-value is less than 0.05, reject null hypothesis.'
      'We have sufficient evidence to say that the mean is not equal between both countries'
      
  )
else:
    print(
       'p-value is larger than 0.05, do not reject null hypothesis.'
       'There is a no significant different between both countries.'
     
  )

```

- Protein quality
- Gender inequality
- Inequality-adjusted income index
- Proportion of population under global poverty line
